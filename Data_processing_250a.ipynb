```python
mat_df_processed = mat_df.copy()
por_df_processed = por_df.copy()

#'yes'/'no' columns
def map_yes_no(df, column_name):
    df[column_name] = df[column_name].map({'no': 0, 'yes': 1})
    return df

# 'sex' col: 'F' -> 0, 'M' -> 1
mat_df_processed['sex'] = mat_df_processed['sex'].map({'F': 0, 'M': 1})
por_df_processed['sex'] = por_df_processed['sex'].map({'F': 0, 'M': 1})

# 'address' col: 'U' (urban) -> 0, 'R' (rural) -> 1
mat_df_processed['address'] = mat_df_processed['address'].map({'U': 0, 'R': 1})
por_df_processed['address'] = por_df_processed['address'].map({'U': 0, 'R': 1})

# 'famsize' col: 'LE3' (less or equal to 3) -> 0, 'GT3' (greater than 3) -> 1
mat_df_processed['famsize'] = mat_df_processed['famsize'].map({'LE3': 0, 'GT3': 1})
por_df_processed['famsize'] = por_df_processed['famsize'].map({'LE3': 0, 'GT3': 1})

# 'Pstatus' col: 'A' (apart) -> 0, 'T' (together) -> 1
mat_df_processed['Pstatus'] = mat_df_processed['Pstatus'].map({'A': 0, 'T': 1})
por_df_processed['Pstatus'] = por_df_processed['Pstatus'].map({'A': 0, 'T': 1})

# 'school' col: 'GP' -> 0, 'MS' -> 1
mat_df_processed['school'] = mat_df_processed['school'].map({'GP': 0, 'MS': 1})
por_df_processed['school'] = por_df_processed['school'].map({'GP': 0, 'MS': 1})

# 'yes'/'no' cols
yes_no_columns = [
    'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
    'higher', 'internet', 'romantic'
]

for col in yes_no_columns:
    mat_df_processed = map_yes_no(mat_df_processed, col)
    por_df_processed = map_yes_no(por_df_processed, col)

display(mat_df_processed.head())
display(por_df_processed.head())
```

```python
print('mat_df_processed binary columns: \n')

binary_columns_to_check = [
    'school', 'sex', 'address', 'famsize', 'Pstatus',
    'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
    'higher', 'internet', 'romantic'
]

for col in binary_columns_to_check:
    if col in mat_df_processed.columns:
        print(f"Unique values for '{col}' in mat_df_processed: ")
        display(mat_df_processed[col].unique())
    else:
        print(f"Column '{col}' not found in mat_df_processed.")

print('por_df_processed binary columns: \n')
for col in binary_columns_to_check:
    if col in por_df_processed.columns:
        print(f"Unique values for '{col}' in por_df_processed: ")
        display(por_df_processed[col].unique())
    else:
        print(f"Column '{col}' not found in por_df_processed.")
```

```python
display(mat_df_processed['age'].unique())
display(mat_df_processed['Medu'].unique())
display(mat_df_processed['Fedu'].unique())
display(mat_df_processed['traveltime'].unique())
display(mat_df_processed['studytime'].unique())
display(mat_df_processed['failures'].unique())
display(mat_df_processed['famrel'].unique())
display(mat_df_processed['freetime'].unique())
display(mat_df_processed['goout'].unique())
display(mat_df_processed['Dalc'].unique())
display(mat_df_processed['Walc'].unique())
display(mat_df_processed['health'].unique())
display(mat_df_processed['absences'].unique())
```

```python
# Ages: 15-17 -> 0 (high school age), 18-22 -> 1 (college/uni age)
age_bins = [14, 17, 22] # Using 14 as lower bound to include 15, 22 as upper bound to include 22
age_labels = [0, 1]

mat_df_processed['age_category'] = pd.cut(mat_df_processed['age'], bins=age_bins, labels=age_labels, right=True, include_lowest=True)
por_df_processed['age_category'] = pd.cut(por_df_processed['age'], bins=age_bins, labels=age_labels, right=True, include_lowest=True)

# Fail (0-9) -> 0, Pass (10-14) -> 1, Good (15-20) -> 2
grade_bins = [-1, 9, 14, 20] # Using -1 to include 0, and 20 for upper bound
grade_labels = [0, 1, 2]

for col in ['G1', 'G2', 'G3']:
    mat_df_processed[f'{col}_category'] = pd.cut(mat_df_processed[col], bins=grade_bins, labels=grade_labels, right=True, include_lowest=True)
    por_df_processed[f'{col}_category'] = pd.cut(por_df_processed[col], bins=grade_bins, labels=grade_labels, right=True, include_lowest=True)

# None (0) -> 0, Low (1-5) -> 1, Medium (6-15) -> 2, High (16+) -> 3
# Using max absences to set the upper bound for 'High'
max_absences = max(mat_df_processed['absences'].max(), por_df_processed['absences'].max())
absences_bins = [-1, 0, 5, 15, max_absences]
absences_labels = [0, 1, 2, 3]

mat_df_processed['absences_category'] = pd.cut(mat_df_processed['absences'], bins=absences_bins, labels=absences_labels, right=True, include_lowest=True)
por_df_processed['absences_category'] = pd.cut(por_df_processed['absences'], bins=absences_bins, labels=absences_labels, right=True, include_lowest=True)

print("mat_df_processed with new categorized columns:")
display(mat_df_processed[['age', 'age_category', 'G1', 'G1_category', 'absences', 'absences_category']].head())

print("\npor_df_processed with new categorized columns:")
display(por_df_processed[['age', 'age_category', 'G1', 'G1_category', 'absences', 'absences_category']].head())
```

```python
# columns to drop (original numeric columns that have been categorized)
columns_to_drop_originals = ['age', 'G1', 'G2', 'G3', 'absences']

original_mat_df_columns = mat_df.columns.tolist()
original_por_df_columns = por_df.columns.tolist()

final_mat_columns_ordered = []
for col in original_mat_df_columns:
    if col in columns_to_drop_originals:
        final_mat_columns_ordered.append(f'{col}_category')
    else:
        final_mat_columns_ordered.append(col)

final_por_columns_ordered = []
for col in original_por_df_columns:
    if col in columns_to_drop_originals:
        final_por_columns_ordered.append(f'{col}_category')
    else:
        final_por_columns_ordered.append(col)

mat_df_final = mat_df_processed[final_mat_columns_ordered].copy()
por_df_final = por_df_processed[final_por_columns_ordered].copy()

print("mat_df_final columns:", len(mat_df_final.columns))
display(mat_df_final.head())

print("\npor_df_final columns:", len(por_df_final.columns))
display(por_df_final.head())
```

```python
nominal_columns = ['Mjob', 'Fjob', 'reason', 'guardian']

# Unify Mjob and Fjob mappings
all_mjob_values = mat_df['Mjob'].unique().tolist() + por_df['Mjob'].unique().tolist()
all_fjob_values = mat_df['Fjob'].unique().tolist() + por_df['Fjob'].unique().tolist()

all_job_types = sorted(list(set(all_mjob_values + all_fjob_values)))
job_mapping = {job: i for i, job in enumerate(all_job_types)}

mat_df_final['Mjob'] = mat_df_final['Mjob'].map(job_mapping)
por_df_final['Mjob'] = por_df_final['Mjob'].map(job_mapping)
mat_df_final['Fjob'] = mat_df_final['Fjob'].map(job_mapping)
por_df_final['Fjob'] = por_df_final['Fjob'].map(job_mapping)

# 'reason' column
all_reason_values = sorted(list(set(mat_df['reason'].unique().tolist() + por_df['reason'].unique().tolist())))
reason_mapping = {reason: i for i, reason in enumerate(all_reason_values)}
mat_df_final['reason'] = mat_df_final['reason'].map(reason_mapping)
por_df_final['reason'] = por_df_final['reason'].map(reason_mapping)

# 'guardian' column
all_guardian_values = sorted(list(set(mat_df['guardian'].unique().tolist() + por_df['guardian'].unique().tolist())))
guardian_mapping = {guardian: i for i, guardian in enumerate(all_guardian_values)}
mat_df_final['guardian'] = mat_df_final['guardian'].map(guardian_mapping)
por_df_final['guardian'] = por_df_final['guardian'].map(guardian_mapping)

print("mat_df_final with encoded nominal columns:")
display(mat_df_final[nominal_columns].head())

print("\npor_df_final with encoded nominal columns:")
display(por_df_final[nominal_columns].head())
```

check students attending both courses and which part could be different or the same

```python
common_cols_for_merge = [
    'school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu',
    'Mjob', 'Fjob', 'reason', 'nursery', 'internet'
]

merged_df = pd.merge(
    mat_df,
    por_df,
    on=common_cols_for_merge,
    how='inner',
    suffixes=('_mat', '_por')
)

display(merged_df.head())
print(len(merged_df))
```

```python
columns_to_compare = []
for col in merged_df.columns:
    if col.endswith('_mat') and col.replace('_mat', '') not in common_cols_for_merge:
        base_col_name = col.replace('_mat', '')
        if f'{base_col_name}_por' in merged_df.columns:
            columns_to_compare.append(base_col_name)
```

```python
any_difference_mask = pd.Series(False, index=merged_df.index)

for col in columns_to_compare:
    mat_col = f'{col}_mat'
    por_col = f'{col}_por'
    if mat_col in merged_df.columns and por_col in merged_df.columns:
        any_difference_mask = any_difference_mask | (merged_df[mat_col] != merged_df[por_col])

num_students_with_diffs = any_difference_mask.sum()
print(f"Number of students with at least one differing variable: {num_students_with_diffs}")

students_with_diffs_df = merged_df[any_difference_mask].copy()

display_cols = []
for col in columns_to_compare:
    display_cols.append(f'{col}_mat')
    display_cols.append(f'{col}_por')

print("\nHead of DataFrame showing students with differences and their differing variables:")
display(students_with_diffs_df[display_cols][:10])
```
